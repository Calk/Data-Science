```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import mnist

%matplotlib inline

# Importing the dataset and showing some values with '1' label
train_values, label_values = mnist.read_mnist_training_set("")
ones_arrays = train_values[label_values==1]
mnist.montage(ones_arrays[:50])
```

```python
from sklearn.decomposition import PCA

# Generating and examining PCA of the data with label 1
pca_1 = PCA(n_components=2)
pca_1.fit(ones_arrays)
X_ones_transformed = pca_1.transform(ones_arrays)

plt.scatter(X_ones_transformed[:,0], X_ones_transformed[:,1], s=20, alpha=0.5)
plt.suptitle('PCA for label 1 - with 2 components')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()

# Generating and examining PCA of the data with label 3
three_arrays = train_values[label_values==3]
pca_3 = PCA(n_components=2)
pca_3.fit(three_arrays)
X_three_transformed = pca_3.transform(three_arrays)

plt.scatter(X_three_transformed[:,0], X_three_transformed[:,1], s=20, alpha=0.5)
plt.suptitle('PCA for label 3 - with 2 components')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()

# There is main a difference between the scatterplots. 
# In the 1 labeled observations there are no points localized at the center of the chart
# In the 3 labeled observations there are points are homogeneous distributed across all the chart
# One possible explanation, is that the pixels localized in the center represents the right and left corner of the image.
# Since the number 1 is usually centered in the image and the number 3 normally is not, the points at the center of the chart
# could represent the pixels at the ledt and right boundaries of the images.

```

```python
# Getting the first five components of PCA for labeled 1 observations and drawing it.
pca_1_5components = PCA(n_components=5)
pca_1_5components.fit(ones_arrays)

mnist.montage(pca_1_5components.components_[:])
```

```python
mnist.montage([pca_1_5components.mean_])
```
