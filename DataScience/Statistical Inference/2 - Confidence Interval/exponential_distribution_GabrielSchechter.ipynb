Sample from the exponential distribution using numpy.random.exponential (note that the scale parameter is 1/λ). Plot a histogram of your samples, and calculate the sample mean and variance.
Extra: calculate the actual mean and variance.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Data
lambda_param = 5
beta = 1.0/lambda_param
exponential_samples = np.random.exponential(beta, 50000)

# Ploting the histogram
plt.hist(exponential_samples,50, ec='blue', lw=2)
plt.show()

# Calculating Sample Mean and Sample Variance
sample_mean = sum(exponential_samples)/len(exponential_samples)
print sample_mean

variance_mean = sum((exponential_samples-sample_mean)**2.0)/len(exponential_samples)
print variance_mean

# Calculation actual mean and variance.
actual_mean = beta
print actual_mean
actual_variance = beta**2
print actual_variance

```

Calculate the MLE for λ and compute it for samples of different sizes. Plot the error of the MLE as a function of the sample size.


```python
# Calculation of MLE for λ for many samples sizes
error_array = []
samples_array = []
sample_sizes = xrange(10,10000,50)
for s in sample_sizes:
    temp_sample = np.random.exponential(beta, s)
    temp_lambda_MLE  = len(temp_sample)/sum(temp_sample)
    temp_error = abs(temp_lambda_MLE - lambda_param)
    error_array.append(temp_error)
    samples_array.append(s)
    
samples_array = np.array(samples_array)
error_array = np.array(error_array)
plt.scatter(samples_array, error_array, s=20, alpha=0.5)
plt.suptitle('Scatter Plot')
plt.xlabel('Samples')
plt.ylabel('Errors')
plt.show()

#Verifying the linear correlation between samples and sum of errors
print np.corrcoef(samples_array, error_array)[0][1]

```
