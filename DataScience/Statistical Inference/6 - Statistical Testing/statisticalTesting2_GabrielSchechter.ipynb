# Question 1
(a) Write a function that receives a vector x of data points and a number $ \mu_0 $, and computes the likelihood ratio test for the null hypothesis $ \mu = \mu_0 $, assuming the data is distributed $\mathcal{N}(\mu, \sigma^2)$ where $\sigma^2$ is unknown. The function should return the likelihood ratio value and its p-value.


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

def likelihood_test(x, muo):
    sample_mean = np.mean(x)
    sample_std_dev = np.std(x)
    
    likelihood_ratio = len(x)*((muo - sample_mean)**2)/(sample_std_dev**2)
    p_value = stats.chisqprob(likelihood_ratio, 1)
    
    return likelihood_ratio, p_value
```

# Question 2
Fix $ \mu $ and $ \mu_0 $ such that $ \mu\ne\mu_0 $

(a) Fix $\sigma$, and repeat the following for different values of n:

- Sample a vector $ X $ of $ n $ samples from $ \mathcal{N}(\mu,\sigma^2) $ and run your function on $ X $.
- Repeat 10 times for each sample size and average results

Plot the p-value as a function of $ n $. 

```python
# Setup parameters
std_dev = 1
muo = 0
mu = 1
sample_sizes = xrange(100)
p_values_final_list = []

for size in sample_sizes:
    p_value_size_list = []
    for it in xrange(10):
        X = np.random.normal(mu, std_dev, size)
        p_value_temp = likelihood_test(X, muo)[1]
        p_value_size_list.append(p_value_temp)
#   Add avg to the global list
    p_values_final_list.append(np.mean(p_value_size_list))

# Plot the ratios chart
plt.scatter(sample_sizes, p_values_final_list, s=20, alpha=0.5)
title = "p-value VS sample_size"
plt.suptitle(title)
plt.xlabel('sample size') 
plt.ylabel("p-values")
plt.show()
```

(b) Fix $ n $, and repeat the following for different values of $ \sigma $:
- Sample a vector $ X $ of $ n $ samples from $ \mathcal{N}(\mu, \sigma^2) $ and run your function on $ X $.
- Repeat 10 times for each $ \sigma $ and average results

Plot the p-value as a function of $ \sigma $.

```python
# Setup parameters
number_of_samples = 1000
muo = 0
mu = 1
std_dev_list = xrange(1,1000)
p_values_final_list = []

for std_dev in std_dev_list:
    p_value_size_list = []
    for it in xrange(10):
        X = np.random.normal(mu, std_dev, number_of_samples)
        p_value_temp = likelihood_test(X, muo)[1]
        p_value_size_list.append(p_value_temp)
#   Add avg to the global list
    p_values_final_list.append(np.mean(p_value_size_list))

# Plot the ratios chart
plt.scatter(std_dev_list, p_values_final_list, s=20, alpha=0.5)
title = "p-value VS std_dev"
plt.suptitle(title)
plt.xlabel('std_dev_values') 
plt.ylabel("p-values")
plt.show()
```
